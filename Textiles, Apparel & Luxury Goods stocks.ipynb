```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_rows', None)
```

```python
data = pd.read_csv('Results_apparel.csv')
```

```python
data_filtered = data.drop(columns=['Price/Earnings (TTM)', 'Price/Sales (TTM)'])
```

```python
data_filtered.head()
```

```python
data_filtered['Revenue Growth Rate Last 3 Years']=data_filtered['Revenue Growth Rate Last 3 Years'].str.rstrip('%')
data_filtered['Revenue Growth Rate Last 3 Years']= data_filtered['Revenue Growth Rate Last 3 Years'].replace(',', '', regex=True)
data_filtered['Revenue Growth Rate Last 3 Years'] = data_filtered['Revenue Growth Rate Last 3 Years'].replace('--',np.nan)
data_filtered['Revenue Growth Rate Last 3 Years'] =  pd.to_numeric(data_filtered['Revenue Growth Rate Last 3 Years'])

data_filtered['Price Change 3 Months']=data_filtered['Price Change 3 Months'].str.rstrip('%')
data_filtered['Price Change 3 Months']= data_filtered['Price Change 3 Months'].replace(',', '', regex=True)
data_filtered['Price Change 3 Months'] = data_filtered['Price Change 3 Months'].replace('--',np.nan)
data_filtered['Price Change 3 Months'] =  pd.to_numeric(data_filtered['Price Change 3 Months'])


data_filtered['Price Change 6 Months']=data_filtered['Price Change 6 Months'].str.rstrip('%')
data_filtered['Price Change 6 Months']= data_filtered['Price Change 6 Months'].replace(',', '', regex=True)
data_filtered['Price Change 6 Months'] = data_filtered['Price Change 6 Months'].replace('--',np.nan)
data_filtered['Price Change 6 Months'] =  pd.to_numeric(data_filtered['Price Change 6 Months'])


data_filtered['Price Change 12 Months']=data_filtered['Price Change 12 Months'].str.rstrip('%')
data_filtered['Price Change 12 Months']= data_filtered['Price Change 12 Months'].replace(',', '', regex=True)
data_filtered['Price Change 12 Months'] = data_filtered['Price Change 12 Months'].replace('--',np.nan)
data_filtered['Price Change 12 Months'] =  pd.to_numeric(data_filtered['Price Change 12 Months'])



data_filtered['Price Change 1 Month']=data_filtered['Price Change 1 Month'].str.rstrip('%')
data_filtered['Price Change 1 Month']= data_filtered['Price Change 1 Month'].replace(',', '', regex=True)
data_filtered['Price Change 1 Month'] = data_filtered['Price Change 1 Month'].replace('--',np.nan)
data_filtered['Price Change 1 Month'] =  pd.to_numeric(data_filtered['Price Change 1 Month'])


data_filtered['Price Change 5 Day']=data_filtered['Price Change 5 Day'].str.rstrip('%')
data_filtered['Price Change 5 Day']= data_filtered['Price Change 5 Day'].replace(',', '', regex=True)
data_filtered['Price Change 5 Day'] = data_filtered['Price Change 5 Day'].replace('--',np.nan)
data_filtered['Price Change 5 Day'] =  pd.to_numeric(data_filtered['Price Change 5 Day'])

```

```python
data_filtered.dropna(subset = ["Description"], inplace=True)
```

```python
print (data_filtered[[ 'Description', 'Price Change 3 Months'] ])
plt.bar(data_filtered['Description'], data_filtered['Price Change 3 Months']) 
plt.rcParams['figure.figsize']= (30,20)
plt.xticks( rotation='vertical')
plt.title('Price Change 3 Months')
plt.show()
```
